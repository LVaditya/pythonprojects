# Data Visualization.
different plots
1.lineplot 2.scatterplot 3. Barplot 4. Histogram 5. Pie chart 6. Boxplot 7.Donut 8.Contour 9. 3d 10. stacked plot

```python
import matplotlib.pyplot as plt
import numpy as np
```

A line chart or line graph, also known as curve chart, is a type of chart which displays information as a series of data points called 'markers' connected by straight line segments.

```python
#line plot
x=[1, 5, 8, 10, 15]
```

```python
y=[10, 20, 30, 20, 50]
```

```python
plt.plot(x,y, linestyle='dotted', marker='*', ms=10, color='r')
```

```python
plt.plot(x, label='sales')
plt.plot(y, label='shows')
plt.grid()
plt.legend()
plt.xlabel('Xs')
plt.ylabel('Ys')
```

```python
plt.subplot(2,1,1)
plt.plot(x,marker='*',color='r')
plt.subplot(2,1,2)
plt.plot(y,marker='o',color='g')
```

A scatterplot is a type of data display that shows the relationship between two numerical variables

```python
#scatterplot
a=np.random.randn(100)
b=np.random.randn(100)
c_list=[i for i in range(100)]
```

```python
a
```

```python
plt.scatter(a,b)
```

```python
plt.scatter(a,b, marker='*', s=100, c=c_list, cmap='viridis')
plt.colorbar()
```

A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent.

```python
#Barplot
x=['A', 'B', 'C', 'D', 'E']
y=[10, 20, 30, 20, 50]
```

```python
plt.bar(x,y, width=0.5, color='y')
```

A histogram is a graph that shows the frequency of numerical data using rectangles. The height of a rectangle (the vertical axis) represents the distribution frequency of a variable (the amount, or how often that variable appears).

```python
y=np.random.randn(1000)
plt.hist(y,bins=50)
```

A pie chart is a circular statistical graphic which is divided into slices to illustrate numerical proportion. In a pie chart, the arc length of each slice is proportional to the quantity it represents.

```python
#Pie Chart
x=[10, 5, 15, 30, 25, 15]
l=['A', 'B', 'C', 'D', 'E', 'F']
```

```python
plt.pie(x,labels=l)
```

To achieve donut plot wedgeprops argument can be used in pie chart. setting the width will enable to adjust the inner radius.

```python
plt.pie(x,labels=l, wedgeprops= dict(width=0.4));
plt.legend()
```

```python
explode_list=[0, 0, 0, 0.1, 0, 0]
plt.pie(x, labels=l, explode=explode_list, shadow=True);#explode function enables the seperation of slices which helps indepth analysis
```

 Boxplot is a method for graphically demonstrating the locality, spread and skewness groups of numerical data through their quartiles.

```python
#boxplot
l1=np.random.normal(60,10,500)
l2=np.random.normal(80,20,500)
l3=np.random.normal(40,20,500)
l4=np.random.normal(40,40,500)
```

```python
plt.boxplot([l1,l2,l3,l4]);
```

```python
plt.boxplot([l1,l2,l3,l4],labels=['f1','f2','f3','f4'],vert=0);
```

```python
#contour plot
x=np.linspace(0,5,50)
y=np.linspace(0,5,50)
X,Y=np.meshgrid(x,y)
```

```python
Z=np.sin(X)**10+np.cos(X*Y+10)+np.cos(X)
```

```python
plt.contour(X,Y,Z,cmap='RdGy')
plt.colorbar()
```

```python
plt.contour(X,Y,Z,cmap='RdGy',alpha=0.5)
```

```python
#3-D plot
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
```

```python
def f(x,y):
  z=np.exp((-(x+2)**2-(y-1)**2)/6)
  return (z)
```

```python
#ax.plot_surface(X,Y,Z,rstride=5,cmap='viridis')
```

```python
X,Y=np.meshgrid(x,y)
Z=f(X,Y)
```

```python
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,Z)
```

```python
ax=plt.axes(projection='3d')
ax.plot_surface(X,Y,Z,rstride=5,cstride=5,cmap='viridis')
```

```python
 #stacked plot
 plt.bar([1, 2, 3, 4, 5], [10, 20, 30, 25, 5])
 plt.bar([1, 2, 3, 4, 5], [1, 5, 12, 15, 13], bottom=[10, 20, 30, 25, 5])
```

```python
plt.barh([1, 2, 3, 4, 5], [10, 20, 30, 25, 5])
plt.barh([1, 2, 3, 4, 5], [1, 5, 12, 15, 13], left=[10, 20, 30, 25, 5])
```

# Tabulate and Table format


```python
from tabulate import tabulate
```

```python
data=[['abc', 100, 50], ['rst', 102, 60], ['mno', 103, 80]]
print(tabulate(data))
```

```python
print(tabulate(data, headers=['Name', 'Roll_no', 'Marks']))
```

```python
#As dictionary
d={'Name':['Alice', 'Bob', 'Charlie', 'Davis', 'Edger', 'Freddy'], 'Age':[24, 19, 27, 34, 32, 29], 'Id_no':[101, 102, 103, 104, 105, 106]}
```

```python
print(tabulate(d, headers='keys'))
```

```python
data=[['Name', 'Roll_no', 'Marks'], ['abc', 100, 50], ['rst', 102, 60], ['mno', 103, 80]]
```

```python
print(tabulate(data, headers='firstrow', showindex= 'always'))
```

```python
print(tabulate(data, headers='firstrow', showindex= 'always', tablefmt='grid'))
```

```python
print(tabulate(data, headers='firstrow', showindex= 'always', tablefmt='github'))
```

```python
print(tabulate(data, headers='firstrow', showindex= 'always', tablefmt='simple'))
```
